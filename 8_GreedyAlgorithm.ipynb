<h2>Greedy Algorithm</h2>
<p>
Write a greedy algorithm for a small security company to calculate the most cost effective employee schedule for 24 hours, preferrably 24/7.  </p>
<ul>
    <li>6 security guards are available</li>
    <li>People working &le; 8 hours will be paid &dollar;15/hr.</li>
    <li>Anyone working overtime, &gt; 8 hours, will be paid an extra <span>&#36;</span>5/hr, totalling <span>&#36;</span>20/hr  </li>
</ul>

```python
import numpy as np
import pandas as pd
import time
import calendar
import datetime 

#Graphics
import matplotlib.pyplot as plt
import matplotlib.style as style
import seaborn as sns
%matplotlib inline
sns.set_style('white')
from IPython.display import HTML
```

```python
#only allow 12 hour days max
money = {x: 15 if x <= 8 else 20 for x in range(1, 13)} 

#Initialize a 1 week schedule with no names or payrates for 6 employees
schedule = {d:[[h, [None, 0]] for h in range(0, 24)] for d in calendar.day_name}
guards = ['Guard A', 'Guard B', 'Guard C', 'Guard D', 'Guard E', 'Guard F']
```

```python
guard = 0 
working = 1 
moneyOwed = 0

for day in schedule.keys():
    for hour in schedule[day]:
        todo = hour[1]
        if todo[0] is None:
            accruingMoney = money[working]
            if accruingMoney > moneyOwed:
                if guard == len(guards)-1:
                    guard = 0
                else:
                    guard += 1
                working = 1
                accruingMoney = money[working]
            todo[0] = guards[guard]
            todo[1] = accruingMoney
            moneyOwed = accruingMoney
            working += 1
            hour = todo
```

```python
results = pd.DataFrame(columns = ['Time','Guard', 'Earned'])
today=1
shift = 0
for d in schedule.keys():
    for hour in schedule[d]:
        guard = hour[1][0]
        money = hour[1][1]
        date = datetime.datetime.strptime('{0} {1} {2}:00'.format(day, today, hour[0]), '%A %d %H:%M')
        results.loc[shift] = [date, guard, money]
        shift += 1
    today += 1
    

results.set_index('Time', inplace=True)
```

<h1>Summary</h1><p>
<h2>Round Robin Scheduler</h2>
This was a simple greedy algorithm with hashtables for the amount of pay and the schedule for the week.  As local calculations were made, the schedule was filled out locally, ultimately ending up in a global solution of minimizing cost and overtime.  If the hourly money earned increased, the controlling loop would only continue and allow overtime pay if all the other guards had already worked their 8 base hours. By loadbalancing in this way, no overtime was earned by any worker.  Because each employee had the same rate of pay, it didn't make sense to redistribute hours in any way. Instead, an employee could work their 8 hours, after which the next employee could start their 8 hours, and so on... Throughout this specific week, half the employees were scheduled 8 more hours than the other half.  This would even out during week 2 though, because the other half of the employees will be scheduled for those 8 missing hours evening things out.  Therefore, number of hours worked would be consistent and similar throughout the employee base.  There are 168 total hours in a week, and all 168 hours were allocated. A reduction in employee base by 3 employees could happen before anyone would start receiving overtime, except during the weeks when there's an extra hour. A remedy for this would be starting the payschedule on Wednesdays, but the workweek on Sundays.  </p>

<h2>Timing</h2>
<p>This sort of fair scheduling technique is called Round Robin and the complexity in the simplest forms is <b> O(1)</b> due to using hash tables for its resource.  When timestamp based approaches are used, it can reach <b>O(log N)</b>, becasue it needs to iterate over timestamps and possibly logs.  Therefore this implementation was likely O(log N) time.  If the employee count was increased, we may see an increase in that time. </p>
